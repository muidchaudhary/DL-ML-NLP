Spiltting Data 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(can.data,can.target)
```

# Model building using MLPClassifier

```python
from sklearn.neural_network import MLPClassifier
```

```python

mlpa = classifier = MLPClassifier(hidden_layer_sizes=(150,100,50), 
                                 max_iter=300,activation = 'relu',
                                 solver='adam',random_state=1) #default iteration 200, default hidden layers 100
```

```python

mlpa.fit(x_train, y_train)
print('Accuracy on the training subset:',(mlpa.score(x_train,y_train)))
print('Accuracy on the test subset:' ,(mlpa.score(x_test, y_test)))
```

```python
y_pred = mlpa.predict(x_test)
y_pred
```

# Evaluating the model

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
mlp_accuracy = accuracy_score(y_test, y_pred)
mlp_classification_report = classification_report(y_test,y_pred)

print(f'accuracy score: {mlp_accuracy}')
print(f'classifcation report : \n  {mlp_classification_report}')
```

# Visualizing results

```python
mlp_cm  = confusion_matrix(y_test,y_pred)
mlp_cm
```

```python
mlp_disp = ConfusionMatrixDisplay(confusion_matrix = mlp_cm)
mlp_disp.plot()
plt.show()
```

# The Model Looks good and performs well !
